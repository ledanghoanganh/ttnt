# Hàm đưa ra bước đi tiếp theo cho trò chơi 8 Puzzle dựa vào ma trận đầu vào dựa vào Simple-Based Reflex Agent

```python
import random

def vi_tri(matrix):
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                return (i, j)

def khop_luat(state, rules):
    return rules[state]

def hanh_dong(rule):
    return random.choice(rule)

def test(ma_tran):
    rules = {
        (0, 0): ["down", "right"],
        (0, 1): ["down", "left", "right"],
        (0, 2): ["down", "left"],
        (1, 0): ["up", "down", "right"],
        (1, 1): ["up", "down", "left", "right"],
        (1, 2): ["up", "down", "left"],
        (2, 0): ["up", "right"],
        (2, 1): ["up", "left", "right"],
        (2, 2): ["up", "left"]
    }
    state = vi_tri(ma_tran)
    rule = khop_luat(state, rules)
    action = hanh_dong(rule)
    return action

if __name__ == "__main__":
    ma_tran = [[0 for i in range(3)] for i in range(3)]
    print("Nhập ma trận trạng thái hiện tại của trò chơi:")
    for i in range(3):
        for j in range (3):
            ma_tran[i][j] = int(input(f"Nhập con số ở vị trí {(i, j)}: "))
    print(f"Nước đi tiếp theo là: {test(ma_tran)}")
```

# Agent AI tự giải trò chơi 8 Puzzle được cho ngẫu nhiên dựa trên Simple-Based Reflect Module

```python
import random

def random_matrix():
    nums = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    matrix = [[0 for i in range(3)] for i in range(3)]
    for i in range(3):
        for j in range(3):
            num = random.choice(nums)
            matrix[i][j] = num
            nums.remove(num)

    return matrix

def vi_tri(matrix):
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                return (i, j)

def khop_luat(state, rules):
    return rules[state]

def hanh_dong(state, rule):
    moves = {
        "up": (-1, 0),
        "down": (+1, 0),
        "left": (0, -1),
        "right": (0, +1)
    }

    choice = random.choice(rule)
    move = moves[choice]

    (i0, j0) = state
    (i1, j1) = (state[0] + move[0], state[1] + move[1])

    temp = matrix[i0][j0]
    matrix[i0][j0] = matrix[i1][j1]
    matrix[i1][j1] = temp

    return choice

def test(ma_tran):
    rules = {
        (0, 0): ["down", "right"],
        (0, 1): ["down", "left", "right"],
        (0, 2): ["down", "left"],
        (1, 0): ["up", "down", "right"],
        (1, 1): ["up", "down", "left", "right"],
        (1, 2): ["up", "down", "left"],
        (2, 0): ["up", "right"],
        (2, 1): ["up", "left", "right"],
        (2, 2): ["up", "left"]
    }

    state = vi_tri(ma_tran)
    rule = khop_luat(state, rules)
    action = hanh_dong(state, rule)

    return action

if __name__ == "__main__":
    matrix = random_matrix()
    print(matrix)

    goal = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]]


    history = set()
    matrix_tuple = tuple(tuple(row) for row in matrix)
    history.add(matrix_tuple)

    res = ""

    flag = True
    while (matrix != goal):
        res += (test(matrix) + "->")
        print(matrix)
        matrix_tuple = tuple(tuple(row) for row in matrix)
        if (matrix_tuple in history):
            flag = False
            break

        history.add(matrix_tuple)

    print(res)
    if flag:
        print(f"Đã giải được {matrix}")
    else:
        print("Không giải được")
```

# AGENT AI cho Robot hút bụi

```python
import random

def random_matrix():
    num = 3
    matrix = [[0 for _ in range(3)] for _ in range(3)]
    index2d = [(i, j) for i in range(3) for j in range(3)]
    
    while num > 0:
        i, j = random.choice(index2d)
        matrix[i][j] = 1
        num -= 1
        index2d.remove((i, j))
    return matrix

def cap_nhat_state(state, action, percept):
    moves = {"up": (-1, 0), "down": (1, 0), "left": (0, -1), "right": (0, 1)}
    if action in moves:
        di, dj = moves[action]
        state["pos"] = (state["pos"][0] + di, state["pos"][1] + dj)
        
    state["visited"].add(state["pos"])
    
    state["has_trash"] = (percept == 1)
    
    if action == "suck":
        state["cleaned_count"] += 1
        state["has_trash"] = False
        
    return state

def khop_luat(state, bando_model):
    if state["has_trash"]:
        return "suck"
        
    valid_moves = bando_model[state["pos"]]
    moves_dict = {"up": (-1, 0), "down": (1, 0), "left": (0, -1), "right": (0, 1)}
    unvisited_moves = []
    
    for move in valid_moves:
        di, dj = moves_dict[move]
        next_pos = (state["pos"][0] + di, state["pos"][1] + dj)
        
        if next_pos not in state["visited"]:
            unvisited_moves.append(move)
            
    if len(unvisited_moves) > 0:
        return random.choice(unvisited_moves)
    else:
        return random.choice(valid_moves)

def hanh_dong(matrix, real_pos, action):
    if action == "suck":
        i, j = real_pos
        matrix[i][j] = 0
        
    elif action in ["up", "down", "left", "right"]:
        moves = {"up": (-1, 0), "down": (1, 0), "left": (0, -1), "right": (0, 1)}
        di, dj = moves[action]
        real_pos = (real_pos[0] + di, real_pos[1] + dj)
        
    return matrix, real_pos

def robot_hut_bui(matrix):
    bando_model = {
        (0, 0): ["down", "right"],
        (0, 1): ["down", "left", "right"],
        (0, 2): ["down", "left"],
        (1, 0): ["up", "down", "right"],
        (1, 1): ["up", "down", "left", "right"],
        (1, 2): ["up", "down", "left"],
        (2, 0): ["up", "right"],
        (2, 1): ["up", "left", "right"],
        (2, 2): ["up", "left"]
    }

    state = {
        "pos": (0, 0),             
        "visited": set([(0, 0)]),  
        "has_trash": False,
        "cleaned_count": 0        
    }
    
    real_pos = (0, 0)  
    last_action = None  
    res = []           
    
    max_steps = 50 
    step = 0
    
    while state["cleaned_count"] < 3 and step < max_steps:
        
        i, j = real_pos
        percept = matrix[i][j]
        
        state = cap_nhat_state(state, last_action, percept)
        
        action = khop_luat(state, bando_model)
        
        matrix, real_pos = hanh_dong(matrix, real_pos, action)
        
        res.append(action)
        last_action = action
        step += 1
        
    return res, matrix

if __name__ == "__main__":
    matrix = random_matrix()
    
    for row in matrix:
        print(row)
        
    hanh_trinh, matrix_sau_don = robot_hut_bui(matrix)
    
    print('->'.join(hanh_trinh))
    
    for row in matrix_sau_don:
        print(row)
```
