# Hàm đưa ra bước đi tiếp theo cho trò chơi 8 Puzzle dựa vào ma trận đầu vào dựa vào Simple-Based Reflex Agent

```python
import random

def vi_tri(matrix):
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                return (i, j)

def khop_luat(state, rules):
    return rules[state]

def hanh_dong(rule):
    return random.choice(rule)

def test(ma_tran):
    rules = {
        (0, 0): ["down", "right"],
        (0, 1): ["down", "left", "right"],
        (0, 2): ["down", "left"],
        (1, 0): ["up", "down", "right"],
        (1, 1): ["up", "down", "left", "right"],
        (1, 2): ["up", "down", "left"],
        (2, 0): ["up", "right"],
        (2, 1): ["up", "left", "right"],
        (2, 2): ["up", "left"]
    }
    state = vi_tri(ma_tran)
    rule = khop_luat(state, rules)
    action = hanh_dong(rule)
    return action

if __name__ == "__main__":
    ma_tran = [[0 for i in range(3)] for i in range(3)]
    print("Nhập ma trận trạng thái hiện tại của trò chơi:")
    for i in range(3):
        for j in range (3):
            ma_tran[i][j] = int(input(f"Nhập con số ở vị trí {(i, j)}: "))
    print(f"Nước đi tiếp theo là: {test(ma_tran)}")
```

# Agent AI tự giải trò chơi 8 Puzzle được cho ngẫu nhiên dựa trên Simple-Based Reflect Module

```python
import random

def random_matrix():
    nums = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    matrix = [[0 for i in range(3)] for i in range(3)]
    for i in range(3):
        for j in range(3):
            num = random.choice(nums)
            matrix[i][j] = num
            nums.remove(num)

    return matrix

def vi_tri(matrix):
    for i in range(3):
        for j in range(3):
            if matrix[i][j] == 0:
                return (i, j)

def khop_luat(state, rules):
    return rules[state]

def hanh_dong(state, rule):
    moves = {
        "up": (-1, 0),
        "down": (+1, 0),
        "left": (0, -1),
        "right": (0, +1)
    }

    choice = random.choice(rule)
    move = moves[choice]

    (i0, j0) = state
    (i1, j1) = (state[0] + move[0], state[1] + move[1])

    temp = matrix[i0][j0]
    matrix[i0][j0] = matrix[i1][j1]
    matrix[i1][j1] = temp

    return choice

def test(ma_tran):
    rules = {
        (0, 0): ["down", "right"],
        (0, 1): ["down", "left", "right"],
        (0, 2): ["down", "left"],
        (1, 0): ["up", "down", "right"],
        (1, 1): ["up", "down", "left", "right"],
        (1, 2): ["up", "down", "left"],
        (2, 0): ["up", "right"],
        (2, 1): ["up", "left", "right"],
        (2, 2): ["up", "left"]
    }

    state = vi_tri(ma_tran)
    rule = khop_luat(state, rules)
    action = hanh_dong(state, rule)

    return action

if __name__ == "__main__":
    matrix = random_matrix()
    print(matrix)

    goal = [[1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]]


    history = set()
    matrix_tuple = tuple(tuple(row) for row in matrix)
    history.add(matrix_tuple)

    res = ""

    flag = True
    while (matrix != goal):
        res += (test(matrix) + "->")
        print(matrix)
        matrix_tuple = tuple(tuple(row) for row in matrix)
        if (matrix_tuple in history):
            flag = False
            break

        history.add(matrix_tuple)

    print(res)
    if flag:
        print(f"Đã giải được {matrix}")
    else:
        print("Không giải được")
```
